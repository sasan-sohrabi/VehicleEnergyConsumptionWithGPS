```python
import pandas as pd
import folium

# Load the data (replace with the path to your Excel or CSV file)
file_path = 'PostData.xlsx'
data = pd.read_excel(file_path)

data
```

```python
# Initialize the map centered on the average latitude and longitude
map_center = [data['lat'].mean(), data['lng'].mean()]
map_ = folium.Map(location=map_center, zoom_start=12)

# Add each location to the map
for _, row in data.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"Info: {row['speed']}"  # Customize with relevant column info if needed
    ).add_to(map_)

# Save the map to an HTML file
map_.save('map_with_points.html')
print("Map created: map_with_points.html")
```
